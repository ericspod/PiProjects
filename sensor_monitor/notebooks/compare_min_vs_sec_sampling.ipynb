```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

def load_csv(filename):
    df=pd.read_csv(filename).dropna()
    df["time"]=pd.to_datetime(df["time"])
    df=df.set_index('time')
    return df
```

```python
min_data=load_csv("logs/sensors_230311_023640.csv")
min_data.keys()
```

```python
sec_data=load_csv("logs/sensors_230507_231802.csv")

grouper=pd.Grouper(freq='60s')
sec_data_mean=sec_data.groupby(grouper).mean()

sec_data_mean.keys()
```

```python
plt.plot(min_data["gas_resistance"])
plt.plot(sec_data_mean["gas_resistance"])
```

```python
k="oxidising"
plt.plot(np.asarray(min_data[k]))
plt.plot(np.asarray(sec_data_mean[k]))
```
